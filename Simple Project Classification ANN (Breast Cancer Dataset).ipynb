# main Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as mis

# Datasets from sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

# Pre-processing

from sklearn.preprocessing import StandardScaler


# Neural Network

from keras import Sequential , layers
```

```python
df = load_breast_cancer()
data = df.data
target = df.target
```

```python
data_frame = pd.DataFrame(df.data, columns = df.feature_names)
data_frame
```

# Data  Pre-Processing 

```python
data_frame['label'] = target
```

```python
data_frame["label"].value_counts()
```

```python
data_frame.groupby('label').mean()
```

```python
plt.figure(figsize=(10,8))
plt.pie([357 , 212] , labels=["1" , "0"] ,colors=["g" , "y"] , explode=[0.1 , 0.1] , autopct="%1.1f%%" )
plt.axis("equal")
plt.title("target" , fontsize=30 , loc="right" , c="m")
plt.show()
```

```python
X = data_frame.drop(columns=["label"] , axis=1)
y = data_frame["label"]
```

```python
# print(X)
```

```python
# print(y)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=123 , shuffle=True)
```

```python
Scaler = StandardScaler()

Scaler.fit_transform(X_train , y_train)
```

# Building the Neural Network

```python
# setting up the layers of Neural Network

model = Sequential([  
                          layers.Flatten(input_shape=(30,)),
                          layers.Dense(20, activation='relu'),
                          layers.Dense(2, activation='sigmoid')
])

model.summary()
```

```python
# compiling the Neural Network

model.compile(optimizer="adam" ,
              loss="sparse_categorical_crossentropy" ,
              metrics=["accuracy"])
```

```python
# training the Neural Network
history = model.fit(X_train , y_train , epochs=15 , validation_split=0.1)
history
```

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print("Accuracy is == " , accuracy)
print("loss is == " , np.round(loss , 4))
```
